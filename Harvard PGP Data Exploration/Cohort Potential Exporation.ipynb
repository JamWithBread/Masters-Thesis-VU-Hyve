```python
demo_ehr_sub.reset_index(drop=True, inplace = True)
demo_ehr_sub.head()
```

```python
# Impute missing values from ehr demographics using survey demographics
```

```python
impute_list = ['Date of Birth', 'Race', 'Gender']
for i, row in demo_ehr_sub.iterrows():
    for col in impute_list:
        profile_id = row['profile_id']
        val = row[col]
        if type(val) == float:
            new_val = demo_surveys_sub[demo_surveys_sub['profile_id'] == profile_id][col].values[0]
            demo_ehr_sub.iloc[i][col] = new_val
```
